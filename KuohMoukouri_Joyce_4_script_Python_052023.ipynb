```python
#Création d'un dataframe contenant les 1à premiers produit à base de champagne.
col = ['foodId', 'label', 'category', 'foodContentsLabel', 'image']
df_aug = pd.DataFrame(columns = col)
```

```python
if response.status_code != 200:
    print("Failure")
else:
    data = response.json()['hints']
    for record in data:
      dic_food = record['food']
      #print(dic_food['foodId'])
      keys = list(dic_food.keys())
      #print(keys)
      keys = [cle for cle in keys if cle in col]
      #print(keys)
      dic_food = {key_: dic_food[key_] for key_ in keys}
      df_aug.loc[len(df_aug), keys] = dic_food
```

```python
df_aug['foodContentsLabel'] = df_aug['foodContentsLabel'].str.upper()
```

```python
df_aug = df_aug.sort_values('image')
df_aug = df_aug.reset_index(drop = True)
```
